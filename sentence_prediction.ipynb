```
def max_length(liste):
 l_max = 0
 for i in liste:
   l_i = len(i)
   if (l_i>l_max):
     l_max = l_i
 return l_max
```

```
def word_to_int(words):
  word_to_dict = {}
  word_to_dict[" "]= 0
  count=1
  for i in range(0,len(words)):
    word = words[i]
    if word in word_to_dict:
      continue;
    else:
      word_to_dict[word]=count
      count =count+1
  return word_to_dict

    
```

```
def int_to_word(encoding):
  dict_to_int = {}
  i=0
  listint = list(encoding.keys())
  while i<len(encoding):
    dict_to_int[i]=listint[i]
    i=i+1
  return dict_to_int

```
